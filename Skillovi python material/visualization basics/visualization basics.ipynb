##### 1.2 Bar plots

1. Bar graph is a graphical representation of data in which we can highlight the category with particular shapes like a rectangle.
2. The length and heights of the bar chart represent the data distributed in the dataset. In a bar chart, we have one axis representing a particular category of a column in the dataset and another axis representing the values associated with it. 
3. Bar charts can be plotted vertically or horizontally. A vertical bar chart is often called a column chart. When we arrange bar charts in a high to low-value counts manner, we called them Pareto charts.

```python
students = ['rohith','Raghav','Shikhar','manoj']  # students who appeared for math exam
marks = [51,87,45,67]  # marks individual scored in exam
plt.bar(students,marks)  # colour needed to fill bars
plt.title(' comparing students marks in bar plot')  
plt.xlabel('students marks in maths')  # labelling the x axis
plt.ylabel('marks')  #labelling y axis
plt.show() #used to show plots
```

```python
#plotting the same bar graph in horizontal way
#we can even add colours to bars, after inputs include color = (desired colour)
students = ['rohith','Raghav','Shikhar','manoj']  # students who appeared for math exam
marks = [51,87,45,67]  # marks individual scored in exam
plt.barh(students,marks,color = 'green')  # colour needed to fill bars
plt.title(' comparing students marks in bar plot')  
plt.xlabel('students marks in maths')  # labelling the x axis
plt.ylabel('marks')  #labelling y axis
plt.show() #used to show plots
```

from above plot we can clearly makout the diffrence marks scored by the diffrent student who all are appeared in exam
 - shikar is the one who scored least in exam
 - raghav is the one who scored highest in exam, manoj and mohith have scored somewhat similar marks

```python
#scatter plot for sepal length and sepal width insights 
fig = iris[iris.Species=='Iris-setosa'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='orange', label='Setosa')
iris[iris.Species=='Iris-versicolor'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='blue', label='versicolor',ax=fig)
iris[iris.Species=='Iris-virginica'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='green', label='virginica', ax=fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length VS Width ")
fig=plt.gcf()
fig.set_size_inches(10,4)
plt.show()
```

##### 1.4 histogram 


1. A histogram is basically used to represent data provided in a form of some groups.
2. A histogram is accurate representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable.

```python
#lets draw a simple histogram with realtime data
#
population_age = [21,53,60,49,25,27,30,42,40,1,2,102,95,8,15,105,70,65,55,70,75,60,52,44,43,42,45]  
bins = [0,10,20,30,40,50,60,70,90,100]  
plt.hist(population_age, bins) # bin represents the range of values that are divided into series of intervals. Bins are generally created of the same size 
plt.xlabel('age groups')  
plt.ylabel('frequency')  
plt.title('Histogram')  
plt.show() 
```

- we have plotted histogram for the data of the different age group of the people with respect to the bin.
- we can observe from the plat that frequency of the people between age 40-50 is greater than any other range

##### 1.5 pie chart 

- A Pie Chart is a circular statistical plot that can display only one series of data.
- The area of the chart is the total percentage of the given data.
- Pie chart slices will be ordered and plotted counter-clockwise:  


```python
# plotting basic pie chart

y = np.array([40, 22, 18,50 ])#no of students in class
mylabels = ["banglore", "channei", "coimbatore", "delhi"]# area wise allocation

plt.pie(y, labels = mylabels)
plt.show()
```

- pie chart is always a circular plot and we can clearly see that the chart is dividing the circle accordingly to the given inputs
- no of students in class areawise is represented in pie where we can observe that the area for delhi is largest followed by banglore, channei and coimbatore

```python
 # plotting basic pie chart with explode 

y = np.array([40, 22, 18,50 ])#no of students in class
mylabels = ["banglore", "channei", "coimbatore", "delhi"]


myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

##### 2.3 A box plot or box-and-whisker plot .

1. A box plot or box-and-whisker plot shows the distribution of data in a way that facilitates comparisons between variables or across levels of a categorical variable.
2. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
# by observing the box plot we can clearly diffrentiate the  flower species with their petal and sepal dimensions respectively
```

- petal length of iris versiculor and iris virginica is overlapping 

```python
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")
#Further, we see that the box plots describe that the setosa usually has smaller features with few outliers.
#The Versicolor species has average features. The virginica species has the longest features widths and lengths as compared to others.
# We'll use jitter=True so that all the points don't fall in single vertical lines
```

##### 2.4 violin plot


- A violin plot depicts distributions of numeric data for one or more groups using density curves.
- On each side of the line is a kernel density estimation to show the distribution shape of the data. Wider sections of the violin plot represent a higher probability that members of the population will take on the given value; the skinnier sections represent a lower probability.
- The violinplot shows density of the length and width in the species. The thinner part denotes that there is less density         whereas the fatter part conveys higher density

```python
# A violin plot combines the benefits of the previous two plots and simplifies them
# Denser regions of the data are fatter, and sparser thiner in a violin plot
sns.violinplot(x="Species", y="PetalLengthCm", data=iris, size=6)
```

 - Thinner part denotes that there is less density, whereas the fatter part conveys higher density
 - from the plot We can see the probability density of the various features.
 - we can have clear picture than the sensity of the iris setosa is diffrent from both the species.

##### 2.5 Facetgrid 

1. This class maps a dataset onto multiple axes arrayed in a grid of rows and columns that correspond to levels of variables in the dataset. The plots it produces are often called “lattice”, “trellis”, or “small-multiple” graphics.
2. It can also represent levels of a third variable with the hue parameter, which plots different subsets of data in different colors
3. creates and visualizes a kernel density estimate of the underlying feature

```python
sns.FacetGrid(iris, hue="Species", size=6) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
# from the plot we can easily make out that, setosa is easily separable while some portions of Versicolor and virginica are mixed.
```

- from the above plot we can clearly make a statement that data distrubution of iris setosa petal length is not overlapping with any other species of iris
- Iris versicolor's petal length is overlapping with the iris virginica 

##### 2.6 Pairplot 

1. Pairplot visualizes given data to find the relationship between them where the variables can be continuous or categorical.
2. Pairplot function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column
3. pairplot shows the bivariate relation between each pair of features

```python
# The diagonal elements in a pairplot show the histogram by default
# We can update these elements to show other things, such as a kde
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3, diag_kind="kde")
```

- pairplot always tries to plot the grid of scatter plots by considering all the variables and histogram by default diagonally
- in above plot we can clearly observe that every single variable is considered for comparing it with each other


## <span style = 'color:blue'> andrew curves</span>.
###### Andrews curves are a method for visualizing multidimensional data by mapping each observation onto a function with mean, distances and variences. We will be getting into andrew curves in future when we are dealing with advanced machine learning algorithms.



1. Named after statistician david F Andrew
2. Andrews Curves are created by defining a finite Fourier Series which is an equation of sine curves. 
3. Andrews curves are used for visualizing high-dimensional data by mapping each observation onto a function. It preserves means, distance, and variances.

```python

from pandas.plotting import andrews_curves
andrews_curves(iris.drop("Id", axis=1), "Species")
```

```python
# connecting the features for each data sample
from pandas.plotting import parallel_coordinates
parallel_coordinates(iris.drop("Id", axis=1), "Species")
```

Another multivariate visualization technique pandas has is radviz Which puts each feature as a point on a 2D plane, and then simulates having each sample attached to those points through a spring weighted, by the relative value for that feature

```python
from pandas.plotting import radviz
radviz(iris.drop("Id", axis=1), "Species")
```
