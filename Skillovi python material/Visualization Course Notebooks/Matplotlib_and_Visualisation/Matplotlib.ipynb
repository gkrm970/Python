# Why Plot?

1. The best way to understand data is through visualisation

2. You can understand trends, outliers and data distribution by running a simple command to plot the data



```python
from numpy.random import randn
x, y = [],[]
max_ = 50
for i in range(max_):
    x.append(i)
    if i == max_//2:
        y.append(11.2131)
    else:
        y.append(randn(1)[0])
    
plt.plot(x,y,'bo')
plt.show()

for i,j in zip(x,y):
    print('{},{}'.format(i,j))
```

# Plot style and color


By default matplotlib connects two points via line but you can change that like almost everything in matplolib!

```python
# Let's start with a simple plot
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0])

plt.axis([-10, 110, -10, 140])

plt.show()
```

You can pass a 3rd argument to change the line's style and color.
For example `"r--"` means "red dashed line".

```python
#Pass a third argument
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0], "r--")

plt.axis([-10, 110, -10, 140])

plt.show()
```

You can also plot multiple lines on one graph very simply: just pass `x1, y1, [style1], x2, y2, [style2], ...`

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r--", [0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g-")

plt.axis([-10, 110, -10, 140])

plt.show()
```

Or simply call `plot` multiple times before calling `show`.

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r--")
plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g-")
plt.axis([-10, 110, -10, 140])
plt.show()
```

You can also draw simple points instead of lines. Here's an example with green dashes, red dotted line and blue triangles.
Check out [the documentation](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot) for the full list of style & color options.

```python
# Plot y=x, y=x**2 and y=x**3 in three different colors and styles

x = np.linspace(-1.4, 1.4, 30)

plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')

plt.show()
```

Make it more fancy but understand that 
1. The plot function returns a list of `Line2D` objects (one for each line).  
2. You can set extra attributes on these lines, such as the line width, the dash style or the alpha level (transparency level).  
3. See the full list of attributes in [the documentation](http://matplotlib.org/users/pyplot_tutorial.html#controlling-line-properties).

```python
x = np.linspace(-1.4, 1.4, 30)

#retrieve the line2D objects
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')

#Set line width for y=x
line1.set_linewidth(3.0)
line1.set_alpha(0.2)

#Set dash style on y=x**2
line1.set_dash_capstyle("round")

#Set transparency level on y=x**3
line3.set_alpha(0.2)
plt.show()
```

# Subplots

1. Lets say you want to show multiple plots in a single image. You can use matplotlib's `subplot` function and arrange the plots in a grid

2. You have to specify the number of `rows` and `columns` in the figure, and the `index` of the subplot you want to draw on (left to right, and top to bottom). 

3. Matplotlib's pyplot keeps track of the currently active subplot (which you can get a reference to by calling `plt.gca()`), so when you call the `plot` function, it draws on the `active` subplot.


```python
x = np.linspace(-1.4, 1.4, 30)

#First plot for y=x
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)

#Second plot for y=x**2
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)

#Third plot for y=x**3
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot = bottow left
plt.plot(x, x**3)

#Fourth plot for y=x**4
#Note that 224 represents (2,2,4) tuple. 
plt.subplot(224)  # 2 rows, 2 columns, 4th subplot = bottom right
plt.plot(x, x**4)
plt.show()
```

You can also create subplots that span across multiple grid cells

```python
# Plot y=x
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot = top left
plt.plot(x, x)

#Plot y=x**2
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot = top right
plt.plot(x, x**2)

#Plot y=x**3
plt.subplot(2, 1, 2)  # 2 rows, *1* column, 2nd subplot = bottom
plt.plot(x, x**3)
plt.show()
```

1. You can also add complex subplot positionning using `subplot2grid` instead of `subplot`

2. For `subplot2grid` also you will have to provide number of `rows` and `columns` and subplot's `position` (top left is still (0,0)!).

3. For position you can specify how many rows and columns the subplot is supposed to span


```python
# Create a 3*3 grid for subplots
plt.subplot2grid((3,3), (0, 0), rowspan=2, colspan=2)    #Spans 2 rows and 2 columns starting with (0,0)
plt.plot(x, x**2)

#Plot y=x**3
plt.subplot2grid((3,3), (0, 2))    #Default spanning(1 row and 1 column) at 0 row and 2nd column
plt.plot(x, x**3)

#Plot y=x**4
plt.subplot2grid((3,3), (1, 2), rowspan=2)    #Span 2 rows and 1 column (default) starting from 1st row and 2nd column
plt.plot(x, x**4)

#Plot y=x**5
plt.subplot2grid((3,3), (2, 0), colspan=2)    #Span 1 row (default) 2 columns starting from 2nd row and 0th column
plt.plot(x, x**5)
plt.show()
```

For more details on subplot, you can refer to the [GridSpec documentation](http://matplotlib.org/users/gridspec.html)

1. Lets say you want to draw multiple plots in one image with each plot containing multiple subplots.

2. By default, matplotlib creates `figure(1)` automatically. When you switch figure, pyplot keeps track of the currently active figure and the active subplot of that figure becomes the current subplot.

```python
x = np.linspace(-1.4, 1.4, 30)

#Plotting first figure
plt.figure(1)
plt.subplot(211)
plt.plot(x, x**2)
plt.title("Square and Cube")
plt.subplot(212)
plt.plot(x, x**3)

#Plotting second figure
plt.figure(2, figsize=(10, 5))
plt.subplot(121)
plt.plot(x, x**4)
plt.title("y = x**4")
plt.subplot(122)
plt.plot(x, x**5)
plt.title("y = x**5")

# Changing first figure's plot style. Note that current subplot of figure 1 is 212
plt.figure(1)
plt.plot(x, -x**3, "r:")

plt.show()
```

# Drawing Texts

1. Let's say you want to add text to your plot to draw user's attention.

2. You can use `text` to add text by specifying `horizontal` and `vertical` coordinates, the text and additional formatting arguments
  

```python
x = np.linspace(-1.5, 1.5, 30)

# Point where you want to draw user's attention
px = 0.8
py = px**2

#Plot the graph with marking point with red dot
plt.plot(x, x**2, "b-", px, py, "ro")


#Add text at (0,1.5) 
plt.text(0, 1.5, "Square function\n$y = x^2$", fontsize=20, color='blue', horizontalalignment="center")

#Add text near the point
plt.text(px - 0.08, py, "Beautiful point", ha="right")    #ha is alias for horizontal alignment

plt.text(px, py, "x = %0.2f\ny = %0.2f"%(px, py), rotation=50, color='gray')

plt.show()
```

If you want to annotate a point, you can use `annotate` function by specifying location of point, text position and formatting arguments for arrow.



```python
plt.plot(x, x**2, px, py, "ro")

plt.annotate("Beautiful point", xy=(px, py), xytext=(px-1.3,py+0.5),
                           color="green", weight="heavy", fontsize=14,
                           arrowprops={"facecolor": "lightgreen"})
plt.show()
```

# Legend and Scales


1. Let's say you want to add `legen` in your plots to make it more understandable. 

2. Add `label` in each of `plot` calls. After you are done call `legend` function. 

```python
x = np.linspace(-1.4, 1.4, 50)

#Add label
plt.plot(x, x**2, "r--", label="Square function")

#Add label
plt.plot(x, x**3, "g-", label="Cube function")

#Call legend
plt.legend(loc="best")

plt.grid(True)
plt.show()
```

1. Let's say you want to draw plots at non-linear scales.

2. You can use `yscales` and `xscales` to specify scales of different axis


```python
x = np.linspace(0.1, 15, 500)
y = x**3/np.exp(2*x)

plt.figure(1)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

plt.figure(2)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

plt.show()
```

# Ticks

Let's say you want to control the markings on `x` and `y` axis, you can use the `tickers` object of matplotlib. Let's first understand the concepts of `ticks`:

1. `Ticks`: This is the *location* of the marks.

2. `Tick Lines`: Small lines drawn near the `ticks`.

3. `Tick Labels`: The labels drawn next to tick lines

4. `Tickers`: This is the object which decide the tick labels, tick lines and ticks. 

Although the default tickers works best but if you want to have more control you can activate monitor ticks and modify the plots

```python
x = np.linspace(-2, 2, 100)

# First plot with default ticks
plt.figure(1, figsize=(15,10))
plt.subplot(131)
plt.plot(x, x**3)
plt.grid(True)
plt.title("Default ticks")

# Second plot with manual ticks
ax = plt.subplot(132)
plt.plot(x, x**3)

#Set ticks manually to [-2,-1,0,1]
ax.xaxis.set_ticks(np.array([-2, -1,  0,  1]))
plt.grid(True)
plt.title("Manual ticks on the x-axis")

#Third plot with monitor ticks
ax = plt.subplot(133)
plt.plot(x, x**3)

#Switch on monitor ticks
plt.minorticks_on()
ax.tick_params(axis='x', which='minor', bottom='off')    # Switch off minor ticks for x-axis
ax.xaxis.set_ticks([-2, 0, 1, 2])
ax.yaxis.set_ticks(np.arange(-5, 5, 1))
ax.yaxis.set_ticklabels(["min", -4, -3, -2, -1, 0, 1, 2, 3, "max"])

plt.title("Manual ticks and tick labels\n(plus minor ticks) on the y-axis")


plt.grid(True)

plt.show()
```
